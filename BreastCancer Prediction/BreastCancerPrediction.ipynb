```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
data_frame.tail()
```

```python
data_frame.isnull().sum()
```

```python
for i in data_frame.columns:
    print(i)
```

```python
data_frame.describe()
```

```python
data_frame['label'].value_counts()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

 <h1 style="color:aqua ;font-size:40px" > Building Neural Network </h1>

```python
import tensorflow as tf
tf.random.set_seed(3) # we will get the same accuracy every time we run this file
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# training the neural network

history=model.fit(X_train_std,Y_train, validation_split=0.1,epochs=50)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['training_data','validation_data'])
plt.show()
```

# Accuracy of the model on test data

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```

```python
Y_pred=model.predict(X_test_std)
```

```python
print(X_test_std)

print(Y_pred)
```

model.predict() gives the prediction probability of each class for that data point

```python
# converting  the prediction probability to class labels

Y_pred_labels=[np.argmax(i) for i in Y_pred] # Gives index of maximum value

print(Y_pred_labels)
```
